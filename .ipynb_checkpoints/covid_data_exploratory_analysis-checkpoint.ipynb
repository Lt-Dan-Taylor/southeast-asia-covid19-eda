<a id='datframe'></a>
### Create the Dataframes ~

We create our dataframes with ``pandas``:
   - As all of our dataset are in ``CSV`` format, let's use pandas `read_csv` function to create the dataframe

```python
covid_df = pd.read_csv('WHO-COVID-19-global-data.csv', parse_dates = ['Date_reported'])

vaccinations_df = pd.read_csv('covid-vaccination-dataset/vaccinations.csv', parse_dates=['date'])

world_data_df = pd.read_csv('countries-of-the-world-2023/world-data-2023.csv')

# Checking all columns name in the all dataframes above
print(' Columns '.center(100, '-'))
print('covid_df:')
print(', '.join(covid_df.columns))
print('')
print('vaccinations_df:')
print(', '.join(vaccinations_df.columns)) 
print('')
print('world_data_df:')
print(', '.join(world_data_df.columns))
```

- Merging our datasets into one dataframe using pandas function ``merge``

```python
# Merging the covid_df and world_data first, then vaccinations_df
merged_df1 = pd.merge(covid_df[['Date_reported', 'Country_code', 'Cumulative_cases', 'New_cases', 'Cumulative_deaths', 'New_deaths']],
                   world_data_df[['Abbreviation', 'Country', 'Life expectancy', 'Physicians per thousand', 'Urban_population', 'Latitude', 'Longitude']],
                   left_on='Country_code', right_on='Abbreviation', how='inner')

merge_df2 = pd.merge(merged_df1, vaccinations_df[['date', 'location', 'people_vaccinated_per_hundred']],
                   left_on=['Country', 'Date_reported'], right_on=['location', 'date'], how='left')

# Print all columns in the dataframe
print('Columns: ', ', '.join(merge_df2.columns))
```


- Although the current column order is functional, it doesn't align with my preferences. The columns name also didn't follow the same format, so let's make some adjustments to it

```python
main_df = merge_df2[['Date_reported', 'Country_code', 'Country', 'New_cases', 'Cumulative_cases', 'New_deaths', 'Cumulative_deaths',
                     'people_vaccinated_per_hundred', 'Physicians per thousand', 'Urban_population', 'Latitude', 'Longitude']].reset_index()

# Define a dictionary mapping old column names to new column names
column_name = {'Date_reported': 'date', 'Country_code': 'country_code', 'Country': 'country_name', 'New_cases': 'daily_new_cases', 
                'Cumulative_cases': 'total_cases', 'New_deaths': 'daily_new_deaths', 'Cumulative_deaths': 'total_deaths', 
                'people_vaccinated_per_hundred': 'vaccination_rate', 'Physicians per thousand': 'physicians_per_thousand', 
                'Urban_population': 'urban_population', 'Latitude': 'latitude', 'Longitude': 'longitude'}

# Rename the columns using the dictionary
main_df.rename(columns=column_name, inplace=True)

print('Columns: ', ', '.join(main_df.columns))
```

<a id='clean'></a>
### Data cleaning and wrangling ~

```python
# Create a function for checking shape, datatypes, and missing value of the dataframe
def dataframe_info(dataframe):
    print(' Shape '.center(60, '-'))
    print(f'{"Columns":<30}{dataframe.shape[1]:>30}')
    print(f'{"Rows":<30}{dataframe.shape[0]:>30}')
    print(' Datatypes '.center(60, '-'))
    for column, dtype in dataframe.dtypes.items():
        print(f'{column:<30}{str(dtype):>30}')
    print(' Missing Values '.center(60, '-'))
    missing_values = pd.isnull(dataframe).sum()
    for column, count in missing_values.items():
        print(f'{column:<30}{count:>30}')

# Pass our dataframe as an argument for the dataframe
dataframe_info(main_df)
```

- As displayed above, the ``Urban_population`` column currently has values stored as text (string-object dataype) instead of numbers. Meanwhile, the ``people_vaccinated_per_hundred`` column shows that some data is missing, with only "52866 non-null" values available
- To work with the data effectively, we need to make a few changes. Firstly by fill in the ``people_vaccinated_per_hundred`` column missing value with previous value or 0 if there isn't any.  And for the ``Urban_population`` column we can start by removing any unwanted characters and then convert it to a datatype ``int64``

```python
main_df.sort_values(by=['country_name', 'date'], inplace=True)

# Fill missing values with forward fill and 0 fill
main_df['vaccination_rate'] = main_df.groupby('country_name')['vaccination_rate'].ffill().fillna(0)
```

```python
# Since Urban_population is a string-object dataype and there are non-numeric characters, we have to get rid of them first
def clean_numeric_value(value):
    if isinstance(value, str):
        return int(value.replace(',', '').replace(' ', '').replace('?', ''))  # Remove commas, spaces and non-numeric 
    return value

#  Apply above function and fill missing value with 0
main_df['urban_population'] = main_df['urban_population'].apply(clean_numeric_value).fillna(0)

# Change the datatype into int64
main_df['urban_population'] = main_df['urban_population'].astype('int64')
```

```python
# Check the dataframe once again
dataframe_info(main_df)
```

<a id='answer'></a>
### Answers and Visualizations ~

1. **What is the overall trend of COVID-19 in Southeast Asia over time?**
   - As to answer this question let's take a look of total and new cases and deaths over time
   - We also need to use Moving Average to see the projection more clearly because it's reducing the noises in our data 

```python
# Here a list of Southeast Asia countries
sea_countries = ['Indonesia', 'Malaysia', 'Singapore', 
                 'Brunei', 'Thailand', 'Cambodia', 
                 'Laos', 'Myanmar', 'Vietnam', 'Philippines']

sea_covid_df = main_df.loc[main_df['country_name'].isin(sea_countries)]

# Checking list of countries in dataframe
sea_covid_df['country_name'].unique()
```

- To look the overall trend in Southeast Asia overtime, we should create a dataframe which consist new and total of both cases and deaths for each ``Date_reported``

```python
# Create the dataframe with groupby
sea_cases_and_deaths_df = sea_covid_df.groupby('date')[['daily_new_cases', 'total_cases', 
                                                        'daily_new_deaths', 'total_deaths']].sum().reset_index()

# Set index to Date_reported
sea_cases_and_deaths_df.set_index('date', inplace=True)

sea_cases_and_deaths_df.sample(3)
```

 - What is the total and all-time high of new cases and deaths?

```python
max_total_cases = sea_cases_and_deaths_df['total_cases'].max()
max_new_cases = sea_cases_and_deaths_df['daily_new_cases'].max()
new_cases_date = sea_cases_and_deaths_df[sea_cases_and_deaths_df['daily_new_cases'] == max_new_cases].index[0].date()

max_total_deaths = sea_cases_and_deaths_df['total_deaths'].max()
max_new_deaths = sea_cases_and_deaths_df['daily_new_deaths'].max()
max_new_deaths_date = sea_cases_and_deaths_df[sea_cases_and_deaths_df['daily_new_deaths'] == max_new_deaths].index[0].date()

print(' Southeast Asia Covid-19 Cases '.center(60, '-'))
print('Total cases:')
print(f'{max_total_cases:,} cases')
print('')
print('All-time high of daily new cases:')
print(f'{max_new_cases:,} new cases a day in {new_cases_date}')
print('')
print(' Southeast Asia Covid-19 Deaths '.center(60, '-'))
print('Total deaths:')
print(f'{max_total_deaths:,} deaths')
print('')
print('All-time high of daily new deaths:')
print(f'{max_new_deaths:,} new deaths a day in {max_new_deaths_date}')
```

- `Matplotlib` and `Seaborn` are the libraries we're going to use for the visualization

```python
from matplotlib.ticker import FuncFormatter

# Define formatter function for the y-axis label
def millions_or_thousands_formatter(x, pos):
    if x >= 1e6:
        return f'{x/1e6:.1f} M'
    elif x >= 1e4:
        return f'{x/1e3:.1f} K'
    else:
        return str(int(x))
        
# Function to create line plots
def line_chart_subplotting(ax, df, data_col, moving_avg_col, label, color):
    sns.lineplot(data=df, x='date', y=data_col, color=color, label=label, ax=ax, linewidth=2)
    sns.lineplot(data=df, x='date', y=moving_avg_col, color='#FFB000', linewidth=2, linestyle='dashed', label=f'Moving Average ({label})', ax=ax)
    ax.tick_params(axis='y', labelsize=10)
    ax.set_ylabel(label, color='black', fontsize=12)
    ax.set_xlabel('Date', color='black', fontsize=12)
    ax.set_ylim(bottom=0)
    ax.yaxis.set_major_formatter(FuncFormatter(millions_or_thousands_formatter))
    ax.legend(loc='upper left')

# Create a figure with subplots
sns.set_style('ticks')
fig, axs = plt.subplots(2, 2, figsize=(13, 7), sharex=True)

# Common window sizes for moving average
window_sizes = {'total_cases': 90,    # 90 days or per quarter
                'daily_new_cases': 7, # 7 days or per week
                'total_deaths': 90,
                'daily_new_deaths': 7}

plot_data = [
    ('total_cases', 'Moving_Average_of_total_cases', 'Total Cases', '#090088'),
    ('daily_new_cases', 'Moving_Average_of_daily_new_cases', 'New Cases', '#2E99B0'),
    ('total_deaths', 'Moving_Average_of_total_deaths', 'Total Deaths', '#C70039'),
    ('daily_new_deaths', 'Moving_Average_of_daily_new_deaths', 'New Deaths', '#EA1179')
]

# Calculate moving averages for each column
for col, window_size in window_sizes.items():
    sea_cases_and_deaths_df[f'Moving_Average_of_{col}'] = sea_cases_and_deaths_df[col].rolling(window=window_size).mean()

# Calculate moving averages for each column and create line plots
for i, (data_col, moving_avg_col, label, color) in enumerate(plot_data):
    ax = axs[i // 2, i % 2]
    line_chart_subplotting(ax, sea_cases_and_deaths_df, data_col, moving_avg_col, label, color)

# Set overall title
fig.suptitle('Southeast Asia Covid-19 Cases and Deaths', fontsize=18)

# tight layout
plt.tight_layout()

# Show the plot
plt.show()
```

2. **How do the COVID-19 trends differ among Southeast Asian countries?**
   - In order to answer this question, let's take a look at the total and the highest daily number of new cases and deaths for each country
   - To visualise this, we'll use an area chart, plotting only the moving average of the given values

- Which country has the most cases and which country has the highest daily new cases? 

```python
country_max = sea_covid_df.groupby('country_name')[['daily_new_cases', 'total_cases', 'daily_new_deaths', 'total_deaths', 'urban_population']].max()

# Total cases and highest daily new cases for each country
country_total_cases = country_max['total_cases'].sort_values(ascending=False)
country_max_new_cases = country_max['daily_new_cases'].sort_values(ascending=False)

# Showing each country total and highest daily new cases sorted by hi
print(' Southeast Asia Covid-19 Cases by Country '.center(60, '-'))
print('Total Cases:')
for country, total_cases in country_total_cases.items():
    print(f'{country:<30}{total_cases:>30,}')
print('')
print('All-time high of daily new cases:')
for country, new_cases in country_max_new_cases.items():
    print(f'{country:<30}{new_cases:>30,}')
```

- Vietnam currently leads Southeast Asia with the highest number of COVID-19 cases, over 11 million, while Cambodia has the lowest, around 138,000 cases in total
- Given the wide range of values, it is difficult to visualise the data effectively. Let's divide the countries into two groups, the top 5 and the bottom 5, both for total cases and for daily new cases


```python
# Let's divide it by two category which are 5 highest and 5 lowest
top5_total_cases = country_total_cases.head(5)
bottom5_total_cases = country_total_cases.tail(5)

# Do it again with daily new cases
top5_new_cases = country_max_new_cases.head(5)
bottom5_new_cases = country_max_new_cases.tail(5)
```

- The area chart is the type of plot that we are going to use for visualisation

```python
# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(13, 8), sharex=True)

# Define colors for each country
country_color = {'Cambodia': '#5B0888', 'Brunei': '#FFCD4B', 'Indonesia': '#FE0000', 'Malaysia': '#38E54D', 'Singapore': '#F1B4BB',
                 'Thailand': '#362FD9', 'Myanmar': '#900C3F', 'Laos': '#6A9C89', 'Vietnam': '#FFB000', 'Philippines': '#EBE4D1'}

# Calculate the maximum values for z-order (which determines if the area value should be in front or behind)
max_new_cases = max(top5_new_cases.max(), bottom5_new_cases.max())
max_total_cases = max(top5_total_cases.max(), bottom5_total_cases.max())

# Create a function to fill the chart areas with z-order based on maximum values
def fill_chart_areas_subplotting(ax, countries, colors, data_column, max_value, label_suffix):
    for i, country in enumerate(countries.index):
        country_data = sea_covid_df[sea_covid_df['country_name'] == country]
        moving_average = country_data[data_column].rolling(window=7).mean()
        ax.fill_between(country_data['date'], 0, moving_average, label=country, color=colors[country], alpha=0.6, zorder=max_value - moving_average.max())

# Fill total cases for the top 5 and bottom 5 countries
fill_chart_areas_subplotting(axs[0, 0], top5_total_cases, country_color, 'total_cases', max_total_cases, 'Total Cases')
fill_chart_areas_subplotting(axs[1, 0], bottom5_total_cases, country_color, 'total_cases', max_total_cases, 'Total Cases')

# Set labels and title for Cumulative_cases
for ax in axs[:, 0]:
    ax.set_ylabel('Total Cases', fontsize=12)
    ax.set_ylim(bottom=0)

axs[0, 0].set_title('Most Covid-19 Cases', fontsize=13)
axs[1, 0].set_title('Fewest Covid-19 Cases', fontsize=13)

# Fill new cases for the top 5 and bottom 5 countries
fill_chart_areas_subplotting(axs[0, 1], top5_new_cases, country_color, 'daily_new_cases', max_new_cases, 'New Cases')
fill_chart_areas_subplotting(axs[1, 1], bottom5_new_cases, country_color, 'daily_new_cases', max_new_cases, 'New Cases')

# Set labels and title
for ax in axs[:, 1]:
    ax.set_ylabel('New Cases', fontsize=12)
    ax.set_ylim(bottom=0)

axs[0, 1].set_title('Highest Covid-19 New Cases', fontsize=13)
axs[1, 1].set_title('Fewest Covid-19 New Cases', fontsize=13)

# Set x-label for the bottom row
axs[1, 0].set_xlabel('Date', fontsize=12)
axs[1, 1].set_xlabel('Date', fontsize=12)

# Move the legend inside the chart for all subplots
for ax in axs.flatten():
    ax.legend(loc='upper left')

# Format y-axis labels as millions or thousands
for ax in axs[:, :2].flatten():
    ax.yaxis.set_major_formatter(FuncFormatter(millions_or_thousands_formatter))

# Set title
fig.suptitle('Southeast Asia Covid-19 Total and New Cases by Country', fontsize=18)

# Show the plot
plt.tight_layout()
plt.show()
```

- Which country has the most deaths by Covid-19, and which has the highest daily new deaths? 

```python
# Total deaths and highest daily new deaths for each country
country_total_deaths = country_max['total_deaths'].sort_values(ascending=False)
country_max_new_deaths = country_max['daily_new_deaths'].sort_values(ascending=False)

# Showing each country total and highest daily new cases sorted by hi
print(' Southeast Asia Covid-19 Deaths by Country '.center(60, '-'))
print('Total Deaths:')
for country, total_deaths in country_total_deaths.items():
    print(f'{country:<30}{total_deaths:>30,}')
print('')
print('All-time high of daily new deaths:')
for country, new_deaths in country_max_new_deaths.items():
    print(f'{country:<30}{new_deaths:>30,}')
```

- As shown above, Indonesia has the highest number of COVID-19 fatalities in Southeast Asia, surpassing 161,900 deaths, while Brunei reports the lowest count at 163 deaths


```python
# Create a figure with a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(13, 8), sharex=True)

# Category for 5 highest and 5 lowest
top5_total_deaths = country_total_deaths.head(5)
bottom5_total_deaths = country_total_deaths.tail(5)
top5_new_deaths = country_max_new_deaths.head(5)
bottom5_new_deaths = country_max_new_deaths.tail(5)

# Calculate the maximum values for z-order (which determines if the area value should be in front or behind)
max_total_deaths = max(top5_total_deaths.max(), bottom5_total_deaths.max())
max_new_deaths = max(top5_new_deaths.max(), bottom5_new_deaths.max())

# Fill total deaths for the top 5 and bottom 5 countries
fill_chart_areas_subplotting(axs[0, 0], top5_total_deaths, country_color, 'total_deaths', max_total_deaths, 'Total Deaths')
fill_chart_areas_subplotting(axs[1, 0], bottom5_total_deaths, country_color, 'total_deaths', max_total_deaths, 'Total Deaths')

# Set labels and title
for ax in axs[:, 0]:
    ax.set_ylabel('Total Deaths', fontsize=12)
axs[0, 0].set_title('Most Covid-19 Deaths', fontsize=13)
axs[1, 0].set_title('Fewest Covid-19 Deaths', fontsize=13)

# Fill new cases for the top 5 and bottom 5 countries
fill_chart_areas_subplotting(axs[0, 1], top5_new_deaths, country_color, 'daily_new_deaths', max_new_deaths, 'New Deaths')
fill_chart_areas_subplotting(axs[1, 1], bottom5_new_deaths, country_color, 'daily_new_deaths', max_new_deaths, 'New Deaths')

# Set labels and title 
for ax in axs[:, 1]:
    ax.set_ylabel('New Deaths', fontsize=12)
axs[0, 1].set_title('Highest Covid-19 New Deaths', fontsize=13)
axs[1, 1].set_title('Fewest Covid-19 New Deaths', fontsize=13)

# Set x-label for the bottom row
axs[1, 0].set_xlabel('Date', fontsize=12)
axs[1, 1].set_xlabel('Date', fontsize=12)

# Move the legend inside the chart for all subplots
for ax in axs.flatten():
    ax.legend(loc='upper left')
    ax.set_ylim(bottom=0)

# Format y-axis labels as millions or thousands
for ax in axs[:, :2].flatten():
    ax.yaxis.set_major_formatter(FuncFormatter(millions_or_thousands_formatter))
    ax.set_ylim(bottom=0)

# Set title
fig.suptitle('Southeast Asia Covid-19 Total and New Deaths by Country', fontsize=18)

# Show the plot
plt.tight_layout()
plt.show()
```

3. **How do Southeast Asian countries differ in terms of infection and severity rates?**
   - Covid-19 spreads primarily through close human contact, with infected individuals releasing tiny respiratory droplets when they cough, sneeze, or talk. Those in close proximity can inhale these droplets, leading to potential transmission. For this reason, I chose to use urban population data for this project, as it better represents the population at greater risk for close contact and potential exposure.
   - To determine the infection rate, we divide the ``total_cases`` by the ``urban_population``. For the severity rate, we calculate it by dividing ``total_deaths`` by ``total cases``

```python
country_max['infection_rate'] = (country_max['total_cases'] / country_max['urban_population']) * 100
country_max['severity_rate'] = (country_max['total_deaths'] / country_max['total_cases']) * 100
```

```python
def bar_chart_subplotting(ax, data, data_column, color, xlabel, ylabel, title):
    sns.set_style('ticks')
    sns.barplot(data=data, x=data.index, y=data_column, ax=ax, color=color)
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    ax.set_title(title, fontsize=14)

# Create a subplot with 2 column and 1 row
fig, axs = plt.subplots(1, 2, figsize=(16, 7))

# Plot infection rate in the first subplot
bar_chart_subplotting(axs[0], country_max, 'infection_rate', '#FFCD4B', '', 'Infection Rate (%)', 'Infection')

# Plot severity rate in the second subplot
bar_chart_subplotting(axs[1], country_max, 'severity_rate', '#B0578D', '', 'Severity Rate (%)', 'Severity')

# Set title for the entire figure
plt.suptitle('Covid-19 Infection and Severity Rates in Southeast Asia', fontsize=18)

# Adjust subplot spacing
plt.subplots_adjust(wspace=3)
plt.tight_layout()


# Show the plot
plt.show()
```
